```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('data/data_science.csv')
df.head()
```

```python
df.shape
```

```python
df.drop(['Unnamed: 0','salary'], axis = 1, inplace = True ) 
```

* axis = 1 : columns are to be dropped. if its rows axis = 0 
* inplace = True means the same data frame is updated withput creating a new. 
* if new_df = df.drop(['Unnamed: 0','salary'], axis=1) then a new df will be created as new_df

```python
df
```

```python
df1 = df.groupby('work_year')['salary_in_usd'].mean().round(2)
df1.plot(kind = 'bar')
```

```python
df1.index
```

```python
df1.values
```

```python
data = {
    'work_year' : df1.index,
    'average_salary':df1.values
}

df1 = pd.DataFrame(data)
df1
```

* pd.DataFrame(): This is a function from the pandas library used to create a DataFrame.* 
data: This is the data you want to populate the DataFrame with. It can be in various forms such as a dictionary, a list of lists, a NumPy array, etc.

```python
df1['average_salary'] = (df1['average_salary']/1000).round(2)
df1
```

```python
ax = df1.plot(kind = 'bar',x = 'work_year' , y = 'average_salary',legend = True)
ax.bar_label(ax.containers[0], labels = df1['average_salary'].map('{:.2f}K'.format))
plt.subplots_adjust(top = 1.2)
ax.legend(['Average Salary'],loc = 'upper left')

plt.xlabel('Work Year')
plt.ylabel('Average Salary')
plt.title('Average Salary by Years', color = 'red')
plt.show()
```

```python
df5 = df[['salary_in_usd','company_size']]
df5
```

```python
S = df5[df5['company_size'] =='S']
M = df5[df5['company_size'] =='M']
L = df5[df5['company_size'] =='L']

labels = ['Medium','Large','Small']

salary_mean = [S['salary_in_usd'].mean(),M['salary_in_usd'].mean(),L['salary_in_usd'].mean()]
salary_mean
```

```python
label_change = np.round([x/1000 for x in salary_mean],2)
label_change = list(map(str,label_change))
label_change = [x + ' K' for x in label_change]
label_change
```

```python
z = plt.bar(labels,salary_mean)
plt.bar_label(z,label_change, alpha = 0.68)
plt.title('Distribution of Salary by company size', color = 'red')
plt.xlabel('Size of Company')
plt.ylabel('Salary in USD')
plt.show()
```

```python
sns.histplot(S['salary_in_usd'],label = 'Small',kde = True)
sns.histplot(M['salary_in_usd'],label = 'Medium',kde = True)
sns.histplot(L['salary_in_usd'],label = 'Large',kde = True)
plt.title('Distribution of Salary by company size', color = 'red')
plt.xlabel('Salary in USD')
plt.show()
```
